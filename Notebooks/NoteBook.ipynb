```python
import pandas as pd
import geopandas as gpd

# 1. Crear GeoDataFrames
gdf_disaster = gpd.GeoDataFrame(global_disaster_df, 
                                 geometry=gpd.points_from_xy(global_disaster_df.longitude, global_disaster_df.latitude), 
                                 crs="EPSG:4326")
gdf_earthquake = gpd.GeoDataFrame(earthquake_df, 
                                   geometry=gpd.points_from_xy(earthquake_df.longitude, earthquake_df.latitude), 
                                   crs="EPSG:4326")

# 2. Proyectar a metros (EPSG:3857) para usar el radio de 50km
gdf_disaster_m = gdf_disaster.to_crs(epsg=3857)
gdf_earthquake_m = gdf_earthquake.to_crs(epsg=3857)

# 3. Crear buffer de 50km en la tabla maestra
gdf_disaster_m['geometry'] = gdf_disaster_m.geometry.buffer(50000)

# 4. Unión espacial
# Esto genera un DataFrame con las columnas de ambos, sufijadas con _left y _right
coincidencias = gpd.sjoin(gdf_earthquake_m, gdf_disaster_m, how='inner', predicate='within')

# 5. FILTRO TEMPORAL: Comparar solo el día (ignora horas/minutos)
# .dt.date extrae la fecha pura de los objetos datetime
duplicados_reales = coincidencias[
    coincidencias['date_left'].dt.date == coincidencias['date_right'].dt.date
]

# 6. Identificar y excluir los duplicados de la tabla de terremotos
# Usamos .index para saber qué filas de earthquake_df "cayeron" en el filtro anterior
earthquake_df_solo_nuevos = earthquake_df[~earthquake_df.index.isin(duplicados_reales.index)]

# 7. Unión final
df_final = pd.concat([global_disaster_df, earthquake_df_solo_nuevos], ignore_index=True)

print(f"Total original global_disaster: {len(global_disaster_df)}")
print(f"Sismos de la segunda tabla descartados por proximidad y misma fecha: {len(duplicados_reales.index.unique())}")
print(f"Total final combinado: {len(df_final)}")
```

```python
for col in df_final.select_dtypes(include=['object', 'string']).columns:
    df_final[col] = df_final[col].astype(str).replace( np.nan, '')

num_cols = df_final.select_dtypes(include=['float64']).columns
df_final[num_cols] = df_final[num_cols].fillna(0)
```

```python
# 1. Definimos la máscara: Terremoto Y (faltan datos en country O subregion O region)
mask = (
    (df_final['disaster_type'] == 'Earthquake') &
    (df_final[['latitude', 'longitude']].notnull().all(axis=1)) &
    (df_final['country'] == '') &
    (df_final['subregion'] == '') &
    (df_final['region'] == '')
)

df_a_procesar = df_final[mask].iloc[0:1000].copy()

respuestas = []

for i in range(len(df_a_procesar)):
    row = df_a_procesar.iloc[i]
    res = mf.call_api(row)
    
    if res:
        respuestas.append(res)
    else:
        # En caso de error crítico, rellenamos con valores de error para mantener el índice
        respuestas.append({'country': 'Error', 'subregion': 'Error', 'region': 'Error'})

# 3. Volcar los datos al DataFrame original
if respuestas:
    print(f"Datos obtenidos para {len(respuestas)} filas.")
    res_df = pd.DataFrame(respuestas)
    # Usamos el índice de df_a_procesar para que coincidan las filas exactamente
    res_df.index = df_a_procesar.index
    
    df_final.loc[res_df.index, ['country', 'subregion', 'region']] = res_df[['country', 'subregion', 'region']]

print("Proceso completado.")
```

```python
df_final['region'] = df_final['region'].replace({'South America':'Americas', 'North America':'Americas', 'Insular Oceania':'Oceania', 'Central America':'Americas', 'Caribbean':'Americas', 'Southern Asia':'Asia', 'Western Asia':'Asia', 'Eastern Asia':'Asia', 'South-Eastern Asia':'Asia', 'Australian continent':'Oceania'})
condicion = (df_final['disaster_type'] == 'Earthquake') & (df_final['magnitude'].astype(float) == 0)
df_final.loc[condicion, 'magnitude'] = np.nan
```

```python
import pymysql
from sqlalchemy import create_engine
from config import user,password,host, bd_name, url


conexion = pymysql.connect(host=host, user=user, password=password)

try:
    with conexion.cursor() as cursor:
        cursor.execute(f"CREATE DATABASE IF NOT EXISTS {bd_name}")
        print(f"Database '{bd_name}' ready.")
finally:
    conexion.close()


engine = create_engine(url)


df_final.to_sql('Natural_disasters', con=engine, if_exists='replace', index=False)

print("Table created succesfully!")
```
